```python
import pandas as pd
import numpy as np
import re
import os
import ast
```

```python
from google.colab import drive
drive.mount('/content/drive')

dataset_path = '/content/drive/MyDrive/DataMan_Songs/Dataset/'
```

```python
## C:\Users\Tommi\Documents\Data Man\Data-Management\Dataset\Info_band.csv
info = pd.read_csv(dataset_path + '3_Info_band.csv')
info = info.rename(columns={'Unnamed: 0': 'Band'})
info = info.replace(np.NaN, '[]')
info['Guitarist'] = info['Guitarist'].apply(lambda x: ast.literal_eval(x))
info['Singer'] = info['Singer'].apply(lambda x: ast.literal_eval(x))
info['Bassist'] = info['Bassist'].apply(lambda x: ast.literal_eval(x))
info['Drummer'] = info['Drummer'].apply(lambda x: ast.literal_eval(x))
info['Pianist'] = info['Pianist'].apply(lambda x: ast.literal_eval(x))

spotify = pd.read_csv(dataset_path + '2_Dataset_SpotifyFeatures.csv')
spotify = spotify.astype(str)


songsterr = pd.read_csv(dataset_path + '1_Dataset_Song_Features_clean.csv')
songsterr = songsterr.astype(str)

## clean songsterr
songsterr['Player'] = songsterr['Player'].astype(str).apply(lambda x: re.sub(r'(Track 1)', 'nan', x))
songsterr['Song'] = songsterr['Song'].astype(str).apply(lambda x: re.sub(r"\((.*?)\)", "", x))
```

```python
print(len(info), len(spotify), len(songsterr))
## Merge songsterr with spotify
df = songsterr.merge(spotify, on = ['Band', 'Song'], how ='right')
print(len(df))

## Check duplicates
df.drop_duplicates(subset=['Band', 'Song'], inplace=True, keep= 'first')
print(len(df))
# df[df['Difficulty'].isnull()]
# df.to_csv(dataset_path + 'Dataset_with_fileaudioTrack.csv', index=False)
```

```python
## Merge df with info band
dt = df.merge(info, on = 'Band', how='left')
dt.columns
dt
```

```python
# Use len(player), len(guitarist)) player when available. Otherside sse the list 
player = dt['Player'].astype(str).to_list()
guitarist = dt['Guitarist'].to_list()

player2 = []
for el in range(len(player)):
    # print(type(player[el]))
    ## if it's nan
    if player[el] == 'nan' or player[el] == 'NaN':
        # print(player[el])
        player[el] = guitarist[el]
    #     # print(type(player[el]), el, guitarist[el])
    else:
        player[el] = [re.sub(r' $', '', (re.sub(r"\((.*?)\)", "", player[el])))]

dt['Player'] = player
dt = dt.drop('Guitarist', axis = 1)
dt['RELEASE_DATE'] = dt['RELEASE_DATE'].apply(lambda x: x[:4])
dt
```
