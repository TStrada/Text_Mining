```python
# IMPORT PACKAGES
import numpy as np
import re
import pandas as pd
```

```python
# IMPORT THE DATASET
df = pd.read_csv('4_DatasetGrafo.csv')
df.columns
df.head(20)
```

```python
pd.options.mode.chained_assignment = None

# preprocessing Instruments_Difficulty column
from ast import literal_eval
def f(x):
    try:
        return literal_eval(str(x))   
    except Exception as e:
        return []
df['Instruments_Difficulty'] = df['Instruments_Difficulty'].apply(lambda x: f(str(x)))

# preprocessing Band and Song columns
def band(x): return ( "'"+ str(x).replace(' ', '_') + "'")
df['Band'] = df['Band'].apply(lambda x: band(x))
df['Song'] = df['Song'].apply(lambda x: band(x))

# preprocessing Singer, Bassist, Drummer and Pianist columns
def singer(x): return(str(x).replace('[','').replace(']','').replace("'",'').replace(',','_'))
df['Singer'] = df['Singer'].apply(lambda x: singer(x))
df['Bassist'] = df['Bassist'].apply(lambda x: singer(x))
df['Drummer'] = df['Drummer'].apply(lambda x: singer(x))
df['Pianist'] = df['Pianist'].apply(lambda x: singer(x))

# preprocessing Technique and Guitarist columns
def technique(x): return(str(x).replace('[','').replace(']','').replace("'",'').replace(',','_').replace(' ',''))
df['Technique'] = df['Technique'].apply(lambda x: technique(x))
df['Guitarist'] = df['Guitarist'].apply(lambda x: technique(x))

pd.options.mode.chained_assignment = None
df.head(10)
```

```python
diz = df.to_dict(orient='records')
diz
```

```python
n_song = 0
couple_nodes = [] 
n_couples = 0       # numero di nodi distinti per strumento+difficoltà (12)
diz_couples = {}
instruments = ['drums', 'guitar', 'bass', 'piano']
```

```python
f = open('FinalGraph.txt','w',encoding='utf-8')
```

```python
for song in diz:
    
    copy1 = {key:song[key] for key in song if (key != 'Pianist') & (key != 'Bassist') &
                                              (key != 'Singer') & (key != 'Guitarist') &
                                              (key != 'Drummer') & (key != 'Player') &
                                              (key != 'Technique') & (key != 'Instruments_Difficulty') &
                                              (key != 'Drums_Energy_Level') & (key != 'Drums_BPM_Level') &
                                              (key != 'Bass_Energy_Level') & (key != 'Bass_BPM_Level') &
                                              (key != 'Piano_Energy_Level') & (key != 'Piano_BPM_Level') &
                                              (key != 'Relevance_Drums') & (key != 'Relevance_Bass') & (key != 'Relevance_Piano')}   
    n_song+=1
    copy1 = re.sub(r'\s','', re.sub(r"'",'',str(copy1)))
    f.write(f"CREATE(track{n_song}:Track" + f"{copy1})\n") 
    
    prova = []
    for couple in song['Instruments_Difficulty']:
        couple = re.sub(r"'",'',couple)
        prova = couple.split('_')
        if couple not in couple_nodes:
            n_couples += 1
            diz_couples[couple] = n_couples
            couple_nodes.append(couple)
            instrument, level = couple.split('_')
            f.write(f"CREATE(cou{diz_couples[couple]}:Instrument_Difficulty" + '{instrument:' + f"'{instrument}'" + f", difficulty:'{level}'" + "})\n")
        if instruments[0] in prova[0]:
            f.write(f"CREATE(cou{diz_couples[couple]})-[:present_in" + '{artist:' + f"'{song['Drummer']}'" + f", relevance:'{song['Relevance_Drums']}'" +'}' + f"]->(track{n_song})\n") 
        if instruments[1] in prova[0]:
            f.write(f"CREATE(cou{diz_couples[couple]})-[:present_in" + '{artist:' + f"'{song['Guitarist']}'" + f", technique:'{song['Technique']}'" +'}' +f"]->(track{n_song})\n")
        if instruments[2] in prova[0]:
            f.write(f"CREATE(cou{diz_couples[couple]})-[:present_in" + '{artist:' + f"'{song['Bassist']}'" + f", relevance:'{song['Relevance_Bass']}'" +'}' + f"]->(track{n_song})\n") 
        if instruments[3] in prova[0]:
            f.write(f"CREATE(cou{diz_couples[couple]})-[:present_in" + '{artist:' + f"'{song['Pianist']}'" + f", relevance:'{song['Relevance_Piano']}'" +'}' + f"]->(track{n_song})\n") 
```

```python
f.close()
```
